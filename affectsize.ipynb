```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

%matplotlib inline
np.random.seed(10) 
```

```python
male_mean=178
male_sd=7.7
```

```python
female_mean=163
female_sd=7.3
```

```python
male_height = stats.norm(male_mean,male_sd)
female_height = stats.norm(female_mean,female_sd)
```

```python
def evaluate_pdf(rv, x=4):
    # evaluating the mean and standard deviation of the random variable
    mean = rv.mean()
    sd = rv.std()
    # using numpy to calculate evenly spaced numbers over a specified interval
    xs = np.linspace(mean - x * sd, mean + x * sd, 200)
    # calculate the pdf values for those xs
    yx = rv.pdf(xs)
    return xs, yx
```

```python
xs, yx = evaluate_pdf(male_height)
plt.plot(xs, yx, label='male', linewidth=4, color='#beaed4') 

# Female height
xs, yx = evaluate_pdf(female_height)
plt.plot(xs, yx, label='female', linewidth=4, color='#fdc086')

# Labels and legend
plt.xlabel('Height (cm)')
plt.legend()
plt.show()
```

```python
male_sample = male_height.rvs(1000)
```

```python
mean1, std1 = male_sample.mean(), male_sample.std()
mean1, std1
```

```python
female_sample = female_height.rvs(1000)
```

```python
mean2, std2 = female_sample.mean(), female_sample.std()
mean2, std2
```

```python
mean_difference = mean1 - mean2
mean_difference
```

```python
relative_difference =mean_difference/mean1
percentage_difference= relative_difference * 100
percentage_difference
```

```python
relative_difference = mean_difference/mean2
percentage_difference= relative_difference * 100
percentage_difference
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

%matplotlib inline
np.random.seed(10) 
```

```python
male_mean=178
male_sd=7.7
```

```python
female_mean=163
female_sd=7.3
```

```python
male_height = stats.norm(male_mean,male_sd)
female_height = stats.norm(female_mean,female_sd)
```

```python
def evaluate_pdf(rv, x=4):
    # evaluating the mean and standard deviation of the random variable
    mean = rv.mean()
    sd = rv.std()
    # using numpy to calculate evenly spaced numbers over a specified interval
    xs = np.linspace(mean - x * sd, mean + x * sd, 200)
    # calculate the pdf values for those xs
    yx = rv.pdf(xs)
    return xs, yx
```

```python
xs, yx = evaluate_pdf(male_height)
plt.plot(xs, yx, label='male', linewidth=4, color="#6c10ef") 

# Female height
xs, yx = evaluate_pdf(female_height)
plt.plot(xs, yx, label='female', linewidth=4, color="#f07e12")

# Labels and legend
plt.xlabel('Height (cm)')
plt.legend()
plt.show()
```

```python
#cohence d
#show step by step on how to get the cohence d
def cohensd (group1, group2):
    difference_mean=group1.mean()-group2.mean()
    n1=len(group1)
    n2=len(group2) 
    v1=group1.var()
    v2=group2.var()
    pooled_variance=(n1*v1 + n2*v2) / (n1+n2)
    pooled_std=np.sqrt(pooled_variance) 
    d= difference_mean/pooled_std 
    return d 
```

```python
#call the function
d=cohensd(male_sample,female_sample) 
d 
```
